```python
import os
import pandas as pd

folder_path = '/content/drive/MyDrive/Pythondataviz/data/2025-01' # Update with the actual path if different

street_files = []
for root, dirs, files in os.walk(folder_path):
  for file in files:
    if '-street' in file and file.endswith(".csv"): # Filter for '-street' in filename and assume CSVs
      file_path = os.path.join(root, file)
      df = pd.read_csv(file_path)
      street_files.append(df)

# Concatenate all DataFrames into a single DataFrame if needed
combined_street_df1 = pd.concat(street_files, ignore_index=True)

# Display the first few rows of the combined DataFrame
if street_files: # Check if the list of files is not empty
    combined_street_df = pd.concat(street_files, ignore_index=True)
    display(combined_street_df.head())
else:
    print("No files containing '-street' were found in the specified folder.")

print(f"Read {len(street_files)} files containing '-street' from {folder_path}")
```
